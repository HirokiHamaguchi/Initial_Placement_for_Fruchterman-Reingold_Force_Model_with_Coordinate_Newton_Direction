```python
import numpy as np
import matplotlib.pyplot as plt


k = 1


def E(d, a):
    return a * (d**3) / (3 * k) - (k**2) * np.log(d)


def gradE(d, a):
    return a * d**2 - k**2 / d


def hessE(d, a):
    return 2 * a * d + k**2 / d**2


print(f"{gradE(0.01,0.01)=}")
print(f"{gradE(0.01,0.1)=}")
print(f"{gradE(0.01,0)=}")

print("")

print(f"{hessE(0.01,0.01)=}")
print(f"{hessE(0.01,0.1)=}")
print(f"{hessE(0.01,0)=}")


for a in [1e-3, 1e-2, 1e-1, 1, 10, 100]:
    x = np.linspace(0.01, 10, 1000)
    y = E(x, a)
    # y = gradE(x, a)
    # y = hessE(x, a)
    plt.plot(x, y, label=f"a={a}")
    plt.scatter(k * pow(a, -1 / 3), E(k * pow(a, -1 / 3), a), c="red")
plt.xlim(0, 10)
plt.ylim(-10, 10)
plt.legend()
plt.show()
```

```python
a = 0.01

x = np.linspace(0.001, 10, 1000)
y = E(x, a)
plt.plot(x, y)
plt.scatter(k * pow(a, -1 / 3), E(k * pow(a, -1 / 3), a), c="red")
plt.xlim(-1, 10)
plt.ylim(-2, 10)

for x0 in [0.01, 0.1, 1, 3]:
    # 2nd order approximation around x0
    y = E(x0, a) + gradE(x0, a) * (x - x0) + 0.5 * hessE(x0, a) * (x - x0) ** 2
    plt.plot(x, y, label=f"x0={x0}")

plt.legend()
plt.show()
```

```python
x = np.linspace(0.01, 5, 1000)
for e in np.linspace(0, 1, 10 + 1):
    y = 1 / np.pow(x, e)
    plt.plot(x, y, label=f"e={e:.1f}")
plt.legend()
plt.ylim(-1, 10)
plt.show()

for e in np.linspace(0, 1, 10 + 1):
    if e != 1:
        y = (np.pow(x, 1 - e) - 1) / (1 - e)
    else:
        y = np.log(x)
    plt.plot(x, y, label=f"e={e:.1f}")
plt.legend()
plt.ylim(-1, 10)
plt.show()
```
